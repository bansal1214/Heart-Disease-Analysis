```python
#Bins of various age categories:

young = df['Age'].quantile(0.25) ##25th quantile of age column

mid_age = df['Age'].median() ##Median Value of Age column

senior = df['Age'].quantile(0.75) ##75th quantile of age column

elder = df['Age'].max() ##Maximum(Highest value) of age
```

```python
## Defining Labels for each bins created

bin_labels = ["Young_age" , "Middle_age" ,"Senior_citizen" , "Elderly"]

bins=[0,young,mid_age,senior,elder]
```

```python
## Creating a new columns 'Age_category' based on bins & Labels

df['Age_Category'] = pd.cut(df['Age'],labels=bin_labels,bins= bins)
```

## Visualization on Age_category column

```python
fig = sns.countplot(x='Age_Category',data = df)

plt.xlabel('Age Category',fontweight ='bold')
plt.ylabel('Count',fontweight='bold')
plt.title('Count of People in Different Age Category' , fontweight='bold')
plt.show()
```

```python
## Adding a new 'Sex_Category' column based on sex column

df['Sex_Category'] = np.where(df['Sex']==1, 'Male', 'Female')

##It assigns 'Male' to the 'Sex_Category' column if the corresponding value in the 'Sex' column is equal to 1, and 'Female' i
```

```python
df
```

```python
#Adding a new column 'ExAng_Category' based on the values in the 'Exercise Induced Angina' column

df['ExAng_Category'] = np.where(df['Exercise Induced Angina'] == 1, 'Yes', 'No')
```

```python
df.head()
```

## Creating Visualization on ChestPain at different Age_category

```python
plt.figure(figsize=(16,9))

fig = sns.countplot(x='ChestPain',hue='Age_Category',data=df)

# Add count labels on top of each bar
for bars in fig.containers:
    fig.bar_label(bars)


plt.xlabel('Chest Pain', fontweight = 'bold')
plt.xticks(rotation = 50)
plt.ylabel('Count', fontweight = 'bold')
plt.title('Chest Pain in Different Age Categories', fontweight = 'bold')

plt.show()

```

```python
#Typical - Characteristic chest pain associated with heart-related issues like a heart attack or angina.

#Asymptomatic - Absence of chest pain, even in the presence of heart-related conditions.

#Nonanginal - Chest pain that is not related to the heart and does not fit the typical symptoms.

#Non-Typical - Chest pain with atypical characteristics, not fitting the classic heart-related symptoms.
```

```python
## Resting BP - BP when person at rest and not engaged in any physical activity or stress-inducing situations.
```

## Creating Visualization on chestPain column with respect to  Resting Blood 

```python
plt.figure(figsize=(10, 6))

fig = sns.boxplot(x = 'ChestPain', y = 'Resting Blood Pressure (mm Hg)' , data = df)

# Calculate the count for each category
count_data = df.groupby(['Resting Blood Pressure (mm Hg)', 'ChestPain']).size().reset_index(name='Count')

plt.xlabel('Chest Pain', fontweight = 'bold')
plt.xticks(rotation = 50)
plt.ylabel('Resting Blood Pressure(mm Hg)', fontweight = 'bold')
plt.title('Chest Pain with respect to Rest BP', fontweight = 'bold')

plt.show()
```

```python
#Visualizations of above data

columns = ['ChestPain','Age_Category','Sex_Category','ExAng_Category']

def plot_cat_count():
    for c in columns :
        plt.figure(figsize=(10, 6))
        fig = sns.countplot(x = c, data = df,palette='dark')
        plt.title(c, fontweight = 'bold')
        plt.ylabel('Count', fontweight = 'bold')
        plt.xlabel(c, fontweight = 'bold')
        plt.tight_layout()
        
        # Add count labels on top of each bar
        for bars in fig.containers:
                fig.bar_label(bars)
       
        plt.show()
        
plot_cat_count()
```

## visualization of Maximum Heart Rate wrt Age Category for different Sex Category

## visualization of Rest BP wrt Sex Category for different Age Category

```python
fig = sns.catplot(x='Sex_Category', y='Resting Blood Pressure (mm Hg)', data=df, col='Age_Category', kind='bar', height=5, aspect=1.2, palette='muted', col_wrap = 2)

plt.suptitle('Resting Blood Pressure in Different Age Categories', fontweight='bold')

plt.tight_layout()

plt.show()
```

```python
# visualization of Cholestrol  wrt Sex Category for different Age Category

fig = sns.catplot(x='Sex_Category', y='Cholestrol (mg/dl)', data=df, col='Age_Category', kind='bar', height=5, aspect=1.2, palette='dark', col_wrap = 2)

plt.suptitle('Cholestrol in Different Age Categories', fontweight='bold')

plt.tight_layout()

plt.show()
```

```python
#Visualization for Relationship Between Cholestrol and RestBP
sns.relplot(x='Cholestrol (mg/dl)', y='Resting Blood Pressure (mm Hg)', data=df, kind='scatter', hue='Sex_Category', col = 'Age_Category', height=6, aspect=1.2, palette='hsv',col_wrap=2)
plt.suptitle('Relationship Between Cholestrol and RestBP', fontweight='bold')
plt.tight_layout()
plt.show()
```
